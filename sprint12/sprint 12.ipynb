## 1.1:  Consulta la API pública JSONPlaceholder utilizando el método GET para obtener:

Lista de publicaciones (/posts)
Lista de usuarios (/users)
Lista de tareas (/todos)

sintaxis general petición -> respuesta JSON -> lista -> DataFrame: 

respuesta = requests.get(f"{url}/endpoint")   # Realiza una petición GET a la API.
datos = respuesta.json()                      # Convierte la respuesta JSON en una lista o diccionario de Python.
df = pd.DataFrame(datos)                      # Convierte los datos en un DataFrame de pandas.
display(df.head())                            # Muestra las 5 primeras filas del DataFrame.

```python
url_1 = "https://jsonplaceholder.typicode.com"

respuesta_posts = requests.get(f"{url_1}/posts")
posts = respuesta_posts.json()
df_posts = pd.DataFrame(posts)
display(df_posts.head())

respuesta_users = requests.get(f"{url_1}/users")
users = respuesta_users.json()
df_users = pd.DataFrame(users)
display(df_users.head())

respuesta_todos = requests.get(f"{url_1}/todos")
todos = respuesta_todos.json()
df_todos = pd.DataFrame(todos)
display(df_todos.head())
```

## 1.2: Muestra por pantalla:

La cantidad total de cada recurso.
El código de estado de cada petición.

```python
print(f"Código de estado: {respuesta_posts.status_code}")                  
print(f"Número de posts: {len(posts)}")                         
print(f"Código de estado: {respuesta_users.status_code}")   
print(f"Número de usuarios: {len(users)}")  
print(f"Código de estado: {respuesta_todos.status_code}") 
print(f"Número de tareas: {len(todos)}")  

```

## 1.4:  Haz una petición POST para crear una nueva publicación ficticia. Incluye el título, el cuerpo del mensaje y un userId.

Muestra la respuesta JSON.
Muestra el código de estado.

```python
nueva_publicacion = {
    "title": "Publicación número 1.",
    "body": "Probando la publicación con POST.",
    "userId": 1
}
respuesta_post = requests.post(f"{url_1}/posts",json=nueva_publicacion)

print(f"Código de estado: {respuesta_post.status_code}")
pprint(respuesta_post.json())
```

## 1.5:  Haz una petición PATCH para modificar parcialmente una publicación existente.

Muestra la respuesta JSON.
Muestra el código de estado.

```python
actualizacion = {"title": "Título actualizado con PATCH"}
respuesta_patch = requests.patch(f"{url_1}/posts/1",json=actualizacion)

print(f"Código de estado: {respuesta_patch.status_code}")
pprint(respuesta_patch.json())
```

## 1.6: Haz una petición DELETE sobre una publicación.

Muestra la respuesta JSON.
Muestra el código de estado.


```python
respuesta_delete = requests.delete(f"{url_1}/posts/1")

print(f"Código de estado: {respuesta_delete.status_code}")
pprint(respuesta_delete.json())
```

## 3.1 Utiliza la API de Open Data BCN para:
Buscar un dataset de tu interés mediante package_search y package_show. .

```python
# url_3 = "https://opendata-ajuntament.barcelona.cat/ca/desenvolupadors#APIS"

url_3 = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search"

parametros_search = {
    "q": "bicicletes"
}

respuesta_search = requests.get(url_3, params=parametros_search)
print("Código de estado:", respuesta_search.status_code)

datos_search = respuesta_search.json()

print("Petición correcta:", datos_search["success"])
print("Datasets encontrados:", datos_search["result"]["count"])

for dataset in datos_search["result"]["results"][:5]:
    print(dataset["title"])

```

## 3.2 De los resultados obtenidos, selecciona uno que tenga recursos disponibles en CSV o JSON.

```python
dataset = datos_search["result"]["results"][0]

url_show = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_show"

parametros_show = {
    "id": dataset["id"]
}

respuesta_show = requests.get(url_show, params=parametros_show)

datos_show = respuesta_show.json()

print("Código de estado:", respuesta_show.status_code)
print("Petición correcta:", datos_show["success"])
print(dataset["title"])

```

```python
for recurso in datos_show["result"]["resources"]:
    print("Nombre:", recurso["name"])
    print("Formato:", recurso["format"])
    print("ID:", recurso["id"])
    print()
```

## 3.3 Con el resource_id del recurso seleccionado, realiza una consulta mediante datastore_search por:

Recuperar al menos 100 registros.

```python
resource_id = "336ba99e-f1df-4359-8f60-4ac3db4927ec"

url_datastore = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search"

parametros_datastore = {
    "resource_id": resource_id,
    "limit": 100
}

respuesta_datastore = requests.get(url_datastore, params=parametros_datastore)

datos_datastore = respuesta_datastore.json()

print("Código de estado:", respuesta_datastore.status_code)
print("Petición correcta:", datos_datastore["success"])
print("Registros recuperados:", len(datos_datastore["result"]["records"]))
```

## 3.4 Convierte los resultados a un DataFrame.

```python
df_bcn = pd.DataFrame(datos_datastore["result"]["records"])
df_bcn.head(2)
```

## 3.5 Convierte el DataFrame en un fichero .csv.

```python
df_bcn.to_csv("bicicletes.csv", index=False)
print("Fichero CSV guardado correctamente.")
```
