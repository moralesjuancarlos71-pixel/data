Calculadora IMC

Este notebook calcula el Índice de Masa Corporal (IMC) a partir del peso y la altura del usuario, validando los datos introducidos y clasificando el resultado según diferentes categorías, utilizaremos dos funciones y las estructuras de control 'while' y 'try-except' .

```python
import re
from collections import Counter
from pprint import pprint
from statistics import mean
from numpy import random
import pyperclip
from pathlib import Path
```

```python
def pedir_datos():
    """
    Solicita al usuario su altura y peso, validando que sean números positivos.

    Returns: tuple: Peso y altura introducidos por el usuario.
    """
    while True:
        try:
            altura = float(input("Introduce tu altura en metros: ").replace(",", "."))
            peso = float(input("Introduce tu peso en kilos: ").replace(",", "."))

            if altura <= 0 or peso <= 0:
                print("Los valores deben ser positivos.")
                continue

            return peso, altura

        except ValueError:
            print("Introduce números válidos.")


def calcular_imc(peso, altura):
    """
    Calcula el índice de masa corporal (IMC).

    Args:
        peso (float): Peso en kilogramos.
        altura (float): Altura en metros.

    Returns: float: IMC redondeado a dos decimales.
    """
    return round(peso / (altura ** 2), 2)


def clasificar_imc(indice_masa_corporal):
    """
    Clasifica el IMC según las categorías establecidas.

    Args: indice_masa_corporal (float): Valor del índice de masa corporal.

    Returns:  str: Categoría correspondiente al IMC.
    """
    if indice_masa_corporal >= 30:
        return "Obesidad"
    elif indice_masa_corporal >= 25:
        return "Sobrepeso"
    elif indice_masa_corporal >= 18.5:
        return "Peso normal"
    else:
        return "Bajo peso"


def principal_imc():
    """
    Ejecuta el programa completo de cálculo del IMC.

    Solicita los datos al usuario, calcula el IMC, obtiene su clasificación y muestra el resultado por pantalla.

    Returns: None.
    """
    peso, altura = pedir_datos()

    indice_masa_corporal = calcular_imc(peso, altura)
    clasificacion = clasificar_imc(indice_masa_corporal)

    print(f"\nTu IMC es {indice_masa_corporal:.2f}")
    print(f"Clasificación: {clasificacion}")

principal_imc()
```

Existen varias unidades de temperatura utilizadas en diferentes contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re).

Convertidor de temperaturas

Este notebook convierte temperaturas entre Celsius, Fahrenheit y Kelvin.

-----  El programa solicita una temperatura y una unidad de origen al usuario, valida que los datos introducidos sean correctos y devuelve todas las conversiones posibles utilizando funciones lambda y diccionarios.

```python
def convertir_temperatura(valor, unidad_origen):
    """
    Convierte temperaturas entre Celsius, Fahrenheit y Kelvin.

    Args:
        valor (float): Temperatura original.
        unidad_origen (str): Unidad de origen.

    Returns: dict , str: Todas las conversiones o mensaje de error.
    """
    conversiones_a_celsius = {
        "celsius": lambda temperatura: temperatura,
        "fahrenheit": lambda temperatura: (temperatura - 32) * 5 / 9,
        "kelvin": lambda temperatura: temperatura - 273.15
    }

    unidad_origen = unidad_origen.lower()

    if unidad_origen not in conversiones_a_celsius:
        return "Unidad no válida"

    celsius = conversiones_a_celsius[unidad_origen](valor)

    conversiones = {
        "celsius": round(celsius, 2),
        "fahrenheit": round(celsius * 9 / 5 + 32, 2),
        "kelvin": round(celsius + 273.15, 2)
    }

    return conversiones


def pedir_datos_temperatura():
    """
    Solicita al usuario la temperatura y la unidad de origen.

    Returns: tuple: Temperatura y unidad introducidas por el usuario.
    """
    while True:
        try:
            temperatura = float(input("Introduce la temperatura: ").replace(",", "."))
            unidad = input("Introduce la unidad (celsius, fahrenheit o kelvin): ").lower()

            return temperatura, unidad

        except ValueError:
            print("Introduce una temperatura válida.")


def principal_temperatura():
    """
    Ejecuta el programa completo de conversión de temperaturas.

    Solicita los datos, convierte la temperatura y muestra el resultado.

    Returns: None.
    """
    temperatura, unidad = pedir_datos_temperatura()
    resultado = convertir_temperatura(temperatura, unidad)

    print(resultado)

principal_temperatura()
```

import re
from collections import Counter
from pprint import pprint
from statistics import mean



```python
def limpiar_texto(texto):
    """
    Limpia el texto eliminando signos de puntuación.

    Args: texto (str): Texto introducido por el usuario.

    Returns: list: Lista de palabras limpias.
    """
    texto_limpio = re.sub(r"[^\w\s]", "", texto)
    return texto_limpio.lower().split()


def contar_palabras(palabras):
    """
    Cuenta cuántas veces aparece cada palabra.

    Args: palabras (list): Lista de palabras.

    Returns: dict: Frecuencia de palabras.
    """
    return dict(Counter(palabras))


def calcular_longitud_media(palabras):
    """
    Calcula la longitud media de las palabras.

    Args: palabras (list): Lista de palabras.

    Returns: float: Longitud media redondeada.
    """
    return round(mean(map(len, palabras)), 2)


def principal_texto():
    """
    Ejecuta el programa completo de análisis de texto.

    Solicita un texto al usuario, limpia las palabras, calcula la frecuenciade palabras y obtiene la longitud media.

    Returns: None.
    """
    texto_usuario = input("Introduce un texto: ")
    palabras = limpiar_texto(texto_usuario)

    if not palabras:
        print("No has introducido palabras.")
        return

    frecuencia_palabras = contar_palabras(palabras)
    longitud_media = calcular_longitud_media(palabras)

    print("Frecuencia de palabras:")
    pprint(frecuencia_palabras)

    print(f"\nLongitud media: {longitud_media}")

principal_texto()
```

Diccionario inverso

Este notebook invierte las claves y valores de un diccionario.

-- Si existen valores duplicados en el diccionario original, el programa agrupa las claves asociadas dentro de una lista para conservar toda la información.

```python
def reverse_dictionary(diccionario):
    """
    Invierte las claves y valores de un diccionario.

    Args: diccionario (dict): Diccionario original.

    Returns: dict: Diccionario con las claves y valores invertidos.
    """
    diccionario_invertido = {}

    for clave, valor in diccionario.items():

        if valor in diccionario_invertido:

            if not isinstance(diccionario_invertido[valor], list):
                diccionario_invertido[valor] = [diccionario_invertido[valor]]

            diccionario_invertido[valor].append(clave)

        else:
            diccionario_invertido[valor] = clave

    return diccionario_invertido


def principal_reverse_dictionary():
    """
    Ejecuta el programa de inversión de un diccionario.

    Crea un diccionario de ejemplo, invierte sus claves y valores y muestra el resultado por pantalla.

    Returns:  None.
    """
    diccionario_frutas = {"x": "apple","y": "banana","z": "banana"}

    resultado = reverse_dictionary(diccionario_frutas)

    print("Diccionario invertido:")
    pprint(resultado)

principal_reverse_dictionary()
```

import re
from collections import Counter
from pprint import pprint

```python
def limpiar_texto(texto):
    """
    Limpia el texto eliminando signos de puntuación.

    Args: texto (str): Texto introducido.

    Returns: list: Lista de palabras limpias.
    """
    texto_limpio = re.sub(r"[^\w\s]", "", texto)
    return texto_limpio.lower().split()


def obtener_palabras(nombre_archivo):
    """
    Lee un archivo TXT y obtiene una lista de palabras limpias.

    Args: nombre_archivo (str): Ruta del archivo.

    Returns: list: Lista de palabras limpias.
    """
    with open(nombre_archivo, "r", encoding="utf-8") as archivo:
        texto = archivo.read()

    return limpiar_texto(texto)


def contar_palabras(palabras):
    """
    Cuenta cuántas veces aparece cada palabra.

    Args: palabras (list): Lista de palabras.

    Returns: dict: Frecuencia de palabras.
    """
    return dict(Counter(palabras))


def agrupar_por_inicial(frecuencias):
    """
    Agrupa palabras según su letra inicial.

    Args: frecuencias (dict): Diccionario de frecuencias.

    Returns: dict: Diccionario agrupado por inicial.
    """
    resultado = {}

    for palabra, frecuencia in sorted(frecuencias.items()):
        inicial = palabra[0]

        if inicial not in resultado:
            resultado[inicial] = {}

        resultado[inicial][palabra] = frecuencia

    return resultado


def principal_palabras_archivo():
    """
    Ejecuta el análisis de palabras de un archivo TXT.

    Lee el archivo, limpia el texto, cuenta las palabras,las agrupa por inicial y muestra el resultado.

    Returns: None.
    """
    # nombre_archivo = r"C:\Users\miren\OneDrive\Datos adjuntos\Documentos\Especialidad IT ACADEMY\sprint 9 Python\TXT tu me quieres blanca.txt"
    nombre_archivo = Path("TXT tu me quieres blanca.txt")



    palabras = obtener_palabras(nombre_archivo)
    frecuencias = contar_palabras(palabras)
    resultado = agrupar_por_inicial(frecuencias)

    pprint(resultado)

principal_palabras_archivo()
```

## 2. Conversión de tipos de datos.

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán todos los elementos que se pudieron convertir en flotantes y la otra donde están los elementos que no se pudieron convertir.

Ejemplo de la lista que recibe el cliente:
lista = ['1.3', 'one','1e10','seven','3-1/2',('2',1,1.4,'not-a-number'),[1,2,'3','3.4']] 

Conversión de tipos de datos

Este notebook analiza una lista con diferentes tipos de datos y separa automáticamente los elementos convertibles y no convertibles a tipo `float`.

El programa recorre estructuras simples y anidadas, como listas y tuplas, utilizando validaciones con `try-except` para evitar errores durante la conversión.

Los resultados se almacenan en dos listas independientes:
- una con los valores convertibles a `float`
- otra con los elementos no convertibles



```python
def conversion(lista):

    convertibles = []
    no_convertibles = []

    for elemento in lista:

        if isinstance(elemento, (list, tuple)):

            for valor in elemento:

                try:
                    convertibles.append(float(valor))

                except ValueError:
                    no_convertibles.append(valor)

        else:

            try:
                convertibles.append(float(elemento))

            except ValueError:
                no_convertibles.append(elemento)

    return convertibles, no_convertibles


lista = ['1.3','one','1e10','seven','3-1/2', ('2', 1, 1.4, 'not-a-number'),[1, 2, '3', '3.4']]
convertibles, no_convertibles = conversion(lista)

print("Los elementos convertibles son:", convertibles)
print("Los elementos no convertibles son:", no_convertibles)
```

from numpy import random
import pyperclip


Este notebook genera contraseñas seguras y personalizables utilizando el módulo `random` de NumPy.

El programa permite definir la longitud de la contraseña y seleccionar distintos tipos de caracteres, como letras mayúsculas, minúsculas, números y signos especiales.

Además, la función valida los parámetros introducidos y garantiza que la contraseña final contenga al menos un carácter de cada grupo seleccionado para mejorar la seguridad.


```python

def crear_contraseña(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):

    """
    Genera una contraseña segura personalizada.

    Args: longitud (int): Longitud de la contraseña.
    mayusculas (bool): Incluye letras mayúsculas.
    minusculas (bool): Incluye letras minúsculas.
    numeros (bool): Incluye números.
    signos (bool): Incluye caracteres especiales.

    Returns: str: Contraseña generada.
    """


    grupo_mayusculas = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    grupo_minusculas = "abcdefghijklmnopqrstuvwxyz"
    grupo_numeros = "0123456789"
    grupo_signos = "!@#$%^&*()-_=+[]{};:,.<>?/|"

    caracteres = ""
    obligatorios = []

    if mayusculas:
        caracteres += grupo_mayusculas
        obligatorios.append(random.choice(list(grupo_mayusculas)))

    if minusculas:
        caracteres += grupo_minusculas
        obligatorios.append(random.choice(list(grupo_minusculas)))

    if numeros:
        caracteres += grupo_numeros
        obligatorios.append(random.choice(list(grupo_numeros)))

    if signos:
        caracteres += grupo_signos
        obligatorios.append(random.choice(list(grupo_signos)))

    if caracteres == "":
        return "Error: debes activar al menos un tipo de carácter"

    if longitud < len(obligatorios):
        return "Error: la longitud es demasiado corta"

    contraseña = obligatorios
    while len(contraseña) < longitud:
        contraseña.append(random.choice(list(caracteres)))

    random.shuffle(contraseña)
    contraseña = "".join(contraseña)

    pyperclip.copy(contraseña)
    return contraseña

resultado = crear_contraseña(10, True, True, True, True)

print(resultado)
print("Contraseña copiada al portapapeles")
```

## 2. Procesamiento de datos simple

Una compañera de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar. Tiene un histórico de partidos de fútbol catalán en un fichero, donde están almacenado los nombres de los equipos y los resultados. Necesita que procesemos los datos de manera automática, para extraer los resultados que necesita.

Utiliza el fichero "historic_partits.txt"

Necesita un programa que devuelva:

El número total de goles que ha hecho cada equipo.
El nombre del equipo más goleador.
El nombre del equipo más goleado
La clasificación global (cada victoria: 3 pts, empate 1 pts, derrota 0 pts)

Este notebook procesa un archivo TXT con resultados históricos de partidos de fútbol.

El programa lee los datos del archivo, calcula los goles a favor y en contra de cada equipo, identifica el equipo más goleador y el más goleado, y genera una clasificación global aplicando 3 puntos por victoria, 1 por empate y 0 por derrota.
from pathlib import Path


```python
def leer_partidos(nombre_archivo):
    """
    Lee un archivo TXT con resultados de partidos.

    Args: nombre_archivo (str): Ruta del archivo.

    Returns: list: Lista de líneas del archivo.
    """
    with open(nombre_archivo, "r", encoding="utf-8") as archivo:
        return archivo.readlines()


def inicializar_equipo(equipos, equipo):
    """
    Inicializa un equipo si no existe en el diccionario.

    Args:
        equipos (dict): Diccionario de equipos.
        equipo (str): Nombre del equipo.

    Returns: dict: Diccionario de equipos actualizado.
    """
    if equipo not in equipos:
        equipos[equipo] = {
            "goles_favor": 0,
            "goles_contra": 0,
            "puntos": 0
        }

    return equipos


def procesar_partidos(lineas):
    """
    Procesa partidos y calcula goles y puntos.

    Args: lineas (list): Lista de partidos.

    Returns: dict: Estadísticas de equipos.
    """
    equipos = {}

    for linea in lineas:
        local, resultado, visitante = linea.strip().split("\t")
        goles_local, goles_visitante = resultado.split("-")

        goles_local = int(goles_local)
        goles_visitante = int(goles_visitante)

        equipos = inicializar_equipo(equipos, local)
        equipos = inicializar_equipo(equipos, visitante)

        equipos[local]["goles_favor"] += goles_local
        equipos[local]["goles_contra"] += goles_visitante
        equipos[visitante]["goles_favor"] += goles_visitante
        equipos[visitante]["goles_contra"] += goles_local

        if goles_local > goles_visitante:
            equipos[local]["puntos"] += 3

        elif goles_local < goles_visitante:
            equipos[visitante]["puntos"] += 3

        else:
            equipos[local]["puntos"] += 1
            equipos[visitante]["puntos"] += 1

    return equipos


def mostrar_resultados(equipos):
    """
    Muestra goles, equipos destacados y clasificación.

    Args: equipos (dict): Estadísticas de equipos.

    Returns: None.
    """
    equipo_mas_goleador = max(
        equipos,
        key=lambda equipo: equipos[equipo]["goles_favor"]
    )

    equipo_mas_goleado = max(
        equipos,
        key=lambda equipo: equipos[equipo]["goles_contra"]
    )

    clasificacion = sorted(
        equipos.items(),
        key=lambda equipo: equipo[1]["puntos"],
        reverse=True
    )

    print("GOLES POR EQUIPO")
    for equipo, datos in equipos.items():
        print(f"{equipo}: {datos['goles_favor']} goles")

    print(f"\nEquipo más goleador: {equipo_mas_goleador}")
    print(f"Equipo más goleado: {equipo_mas_goleado}")

    print("\nCLASIFICACIÓN")
    for posicion, (equipo, datos) in enumerate(clasificacion, start=1):
        print(f"{posicion}. {equipo}: {datos['puntos']} puntos")


def principal_partidos():
    """
    Ejecuta el programa completo de análisis de partidos.

    Lee el archivo, procesa los resultados y muestrala clasificación final.

    Returns: None.
    """
    
    ruta = Path("TXT historic partits.txt")
    
    lineas = leer_partidos(ruta)
    equipos = procesar_partidos(lineas)

    mostrar_resultados(equipos)

principal_partidos()
```
